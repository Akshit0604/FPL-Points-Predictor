```python
import pandas as pd
from google.colab import drive
#drive.mount('/content/drive')

#df= pd.read_csv('/content/drive/My Drive/TDL_Project/FPL_Input.csv')
df= pd.read_csv('FPL_Data_GW29.csv')
print(df.columns)
df
```

```python
#feature engineering
#commented rows need to be added correctly

#df['avg_pts_last_five']= df.groupby('name')['total_points'].rolling(window=5, min_periods=1).mean().reset_index(drop=True)
#df['avg_mins_last_five']= df.groupby('name')['minutes'].rolling(window=5, min_periods=1).mean().reset_index(drop=True)
#df['opponent_goals_conceded'] = df.apply(lambda row: df.loc[df['team_name'] == row['opponent_team_name'], 'goals_conceded'].sum(), axis=1)
#df['opponent_goals_scored'] = df.apply(lambda row: df.loc[df['team_name'] == row['opponent_team_name'], 'goals_scored'].sum(), axis=1)
#df['avg_pts_by_pos_against_opponent']=

df['total_pts']= df.groupby('name')['total_points'].transform('sum')
df['num_occurrences']= df.groupby('name')['name'].transform('count')
df['total_mins']= df.groupby('name')['minutes'].transform('sum')
df['pts_per_min']=df['total_pts']/df['total_mins']
df['avg_pts']= df['total_pts'] / df['num_occurrences']
df['total_goals']= df.groupby('name')['goals_scored'].transform('sum')
df['total_assists']= df.groupby('name')['assists'].transform('sum')
df['total_cs']= df.groupby('name')['clean_sheets'].transform('sum')
df['goals_per_min']=df['total_goals']/df['total_mins']
df['assists_per_min']=df['total_assists']/df['total_mins']
df['cs_per_min']=df['total_cs']/df['total_mins']

df=df[['name', 'team_name', 'opponent_team_name', 'was_home', 'foot', 'sub_position', 'height_in_cm', 'total_points', 'round', 'total_pts', 'pts_per_min', 'avg_pts', 'total_mins', 'total_goals', 'total_assists', 'total_cs', 'goals_per_min', 'assists_per_min', 'cs_per_min']]

df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

#split data into train and test (test is round 29)
test_df= df[pd.isnull(df['total_points'])]
df= df[pd.notnull(df['total_points'])]

# Define categorical columns
categorical_columns = ['name', 'team_name', 'opponent_team_name', 'was_home', 'foot', 'sub_position']
numerical_columns= ['height_in_cm', 'round', 'total_pts', 'pts_per_min', 'avg_pts', 'total_mins', 'total_goals', 'total_assists', 'total_cs', 'goals_per_min', 'assists_per_min', 'cs_per_min']


# Create OneHotEncoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')

#normalising numerical columns
scaler = MinMaxScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Split data into features and target
df_copy = df.copy()
X = df_copy[categorical_columns + numerical_columns]
y = df_copy['total_points']

# Separate categorical and numerical features
categorical_features = X[categorical_columns]
numerical_features = X[numerical_columns]

# Encode categorical features
encoded_features = encoder.fit_transform(categorical_features)

# Combine encoded features with numerical features
X = pd.concat([pd.DataFrame(encoded_features), numerical_features], axis=1)

X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Reshape data for LSTM
X_train = X.values.reshape((X.shape[0], 1, X.shape[1]))

# Split data into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y, test_size=0.2, random_state=42)

# Create an LSTM model
model = Sequential()
model.add(LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_valid, y_valid))

# Evaluate the model
y_pred = model.predict(X_valid)
mae = mean_absolute_error(y_valid, y_pred)
print(f"MAE: {mae:.2f}")
```

```python
result_df= test_df.copy()

test_df[numerical_columns] = scaler.transform(test_df[numerical_columns])
test_categorical_features = test_df[categorical_columns]
test_numerical_features = test_df[numerical_columns]
test_encoded_features = encoder.transform(test_categorical_features)
test_X = pd.concat([pd.DataFrame(test_encoded_features, index=test_numerical_features.index), test_numerical_features], axis=1)
test_X = test_X.values.reshape((test_X.shape[0], 1, test_X.shape[1]))
test_y_pred = model.predict(test_X)

result_df['predicted_points']=test_y_pred
result_df.loc[:, ['name', 'team_name', 'opponent_team_name', 'was_home', 'predicted_points', 'sub_position', 'round']].sort_values(by='predicted_points', ascending=False)
```
