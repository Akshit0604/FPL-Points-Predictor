```python
import pandas as pd
import requests

base_url= 'https://fantasy.premierleague.com/api/element-summary/'

data= []

for i in range (1,705):
    url= base_url+ str(i)+ '/'
    response = requests.get(url)
    if response.ok:
        df= pd.read_json(url, orient='index')
        df= pd.json_normalize(response.json())
        data.append(df)
    else:
        continue
data=pd.concat(data)
```

```python
history= data.iloc[:,1]
history_df=pd.json_normalize(history)

col_list = []
for i in range(32): #change for i range to latest completed gw
    col_list.append(history_df[i])
player_history_df= pd.concat(col_list, axis=0)
player_history_df=pd.json_normalize(player_history_df)
player_history_df.drop(player_history_df.columns[0], axis=1)
player_history_df = player_history_df.rename(columns={player_history_df.columns[0]: 'player_id'})

mapping_df= pd.read_csv('player_id_names_mapper.csv')
player_history_df= pd.merge(player_history_df, mapping_df, on='player_id', how='left')
team_name_mapper = mapping_df.set_index("team_id")["team_name"].to_dict()
player_history_df["opponent_team_name"] = player_history_df["opponent_team"].map(team_name_mapper)

player_history_df
```

```python
column_names = player_history_df.columns.tolist()
new_column_order = column_names[:1] + column_names[-7:] + column_names[1:-7]
player_history_df = player_history_df[new_column_order]
player_history_df = player_history_df.drop(['fixture', 'kickoff_time'], axis=1)

#player_history_df.to_csv('test.csv')
player_history_df
```

```python
fixtures_df= pd.read_csv('fixtures.csv')
fixtures_df=fixtures_df.iloc[:, [1, 9, 11]]
#*** try to add difficulty rating cols 14&15 as parameters later***
fixtures_df.rename(columns={'event': 'round'}, inplace=True)
fixtures_df = fixtures_df.rename(columns={'team_a': 'away_team_id', 'team_h': 'home_team_id'})

print(fixtures_df)
# Merge fixtures_df with mapping_df on away_team_id and team_id
fixtures_df= fixtures_df.merge(mapping_df, left_on='away_team_id', right_on='team_id')

# Rename the team_name column to away_team_name
fixtures_df= fixtures_df.rename(columns={'team_name': 'away_team_name'})

# Drop the team_id column as it is redundant
fixtures_df= fixtures_df.drop(columns='team_id')

# Repeat the same steps for home_team_id and team_id
fixtures_df= fixtures_df.merge(mapping_df, left_on='home_team_id', right_on='team_id')
fixtures_df= fixtures_df.rename(columns={'team_name': 'home_team_name'})
fixtures_df= fixtures_df.drop(columns='team_id')
fixtures_df= fixtures_df.iloc[:, [0, 1, 2, 5, 11]]
fixtures_df = fixtures_df.drop_duplicates()
print(len(fixtures_df))
fixtures_df
```

```python
final_df = player_history_df.copy()
dfs = []

for index, row in fixtures_df.iterrows():

  team_id = row['away_team_id']
  team_name = row['away_team_name']
  opponent_team = row['home_team_id']
  opponent_team_name = row['home_team_name']

  players = player_history_df[player_history_df['team_id'] == team_id][final_df.columns]

  players['round'] = row['round']
  players['opponent_team'] = opponent_team
  players['opponent_team_name'] = opponent_team_name
  players['was_home'] = False

  dfs.append(players)

  team_id = row['home_team_id']
  team_name = row['home_team_name']
  opponent_team = row['away_team_id']
  opponent_team_name = row['away_team_name']

  players = player_history_df[player_history_df['team_id'] == team_id][final_df.columns]
  players['round'] = row['round']
  players['opponent_team'] = opponent_team
  players['opponent_team_name'] = opponent_team_name
  players['was_home'] = True
  dfs.append(players)

dfs=pd.concat(dfs)
final_df = pd.concat((final_df, dfs), ignore_index=True)

final_df = final_df.sort_values(by=['round', 'team_id'], ascending=[True, True])
final_df = final_df.drop_duplicates(subset=['round', 'player_id'])

final_df
```

```python
import numpy as np
last_round=player_history_df['round'].max()

for index, row in final_df.iterrows():

  if(row['round']>last_round):

    final_df.loc[index, ['total_points' , 'team_h_score', 'team_a_score', 'minutes', 'goals_scored',
       'assists', 'clean_sheets', 'goals_conceded', 'own_goals', 'penalties_saved', 'penalties_missed', 'yellow_cards', 'red_cards',
       'saves', 'bonus', 'bps', 'influence', 'creativity', 'threat', 'ict_index', 'starts', 'expected_goals', 'expected_assists',
       'expected_goal_involvements', 'expected_goals_conceded', 'value', 'transfers_balance', 'selected', 'transfers_in', 'transfers_out']]= np.nan

final_df
```
